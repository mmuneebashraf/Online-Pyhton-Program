# 1. Make a calculator using Python with addition , subtraction , multiplication ,division and power.


```python
def add(a,b): 
    c = a+b
    print(c)

def sub(a,b): 
    c = a-b
    print(c)
    
def mul(a,b): 
    c = a*b
    print(c)
    
def div(a,b): 
    c = a/b
    print(c)
    
def pow(a,b): 
    c = a**b
    print(c)

add(3,6)
sub(9,5)
mul(5,4)
div(16,4)
pow(3,2)
```

# 4. Write a Python program to sum all the numeric items in a dictionary.

```python
x={'a':10,'b':20,'c':30,'d':40,'e':50}
y = 0
print(x.values())

for i in x:
    y= y+x[i]
print("Sum of all Numeric Values is: ",y)
```
