```python
eng = 78;
isl = 89;
math = 97;
totalMarks = 300;
totalMarksInPercentage = ((eng + isl + math) / totalMarks) * 100;

print ("You Have Got:" , totalMarksInPercentage, "% Marks")

if (totalMarksInPercentage >=90 and totalMarksInPercentage <=100):
    print ("You Got Grade A+");
elif (totalMarksInPercentage >=80 and totalMarksInPercentage <=89):
    print("You Got Grade A");
elif (totalMarksInPercentage >=70 and totalMarksInPercentage <=79):
    print ("You Got Grade B");
elif (totalMarksInPercentage <70 and totalMarksInPercentage >=0):
    print(" Sorry Dear No Grades For This Time Better Luck Next Time");
else:
    print ("You Made Some Mistake Please Enter Correct Marks Between 0 to 10");
    
```
